```python
unique_diseases_vata_list = unique_diseases_vata_df[['Unique_vata_diseases']].values.tolist()
# flatten list 
unique_diseases_vata_list = [element for innerList in unique_diseases_vata_list for element in innerList]

# unique_diseases_vata_list
```

```python
def flatten_list(input_list):
    """
    Flattens a list of lists into one long, flat list.
    """
    input_list = [element for innerList in input_list for element in innerList]
    return input_list

def remove_duplicates_from_list(x):
  return list(dict.fromkeys(x))
```

```python
# creating hpid list corresponding to the unique vata diseases

hpid_list_corresponding_to_unique_vata_diseases = []

for disease in unique_diseases_vata_list:
    hpid_list_corresponding_to_unique_vata_diseases.append(disease_hpo_dict[disease])

hpid_list_corresponding_to_unique_vata_diseases = flatten_list(hpid_list_corresponding_to_unique_vata_diseases)
hpid_list_corresponding_to_unique_vata_diseases = remove_duplicates_from_list(hpid_list_corresponding_to_unique_vata_diseases)
len(hpid_list_corresponding_to_unique_vata_diseases)
```

```python
# creating the matrix 

unique_diseases_hpid_df = pd.DataFrame(columns= unique_diseases_vata_list, index= hpid_list_corresponding_to_unique_vata_diseases)
```

```python
unique_diseases_vata_set = set(unique_diseases_vata_list)
type(unique_diseases_vata_set)
```

```python
for x in hpid_list_corresponding_to_unique_vata_diseases:
    temp_disease_list = hpo_disease_dict[x]
    temp_disease_set = set(temp_disease_list)

    res_disease_set = temp_disease_set & unique_diseases_vata_set

    for y in res_disease_set:
        unique_diseases_hpid_df[y][x] = 1

unique_diseases_hpid_df = unique_diseases_hpid_df.fillna(0)
unique_diseases_vata_df
```
