```python
def calculte_hpid_freq(phenotype_df, hpo_disease_dict):
    """
    Calculate the frequency of each HPO term in a phenotype.
    """

    retrun_frequecy_dict = defaultdict(int)

    for i in range(len(phenotype_df)):
        # print(i)
        # print(phenotype_df.iloc[i, 0])
        temp_frequency = len(hpo_disease_dict[phenotype_df.iloc[i, 0]])
        # if ()
        retrun_frequecy_dict[phenotype_df.iloc[i, 0]] = temp_frequency

    return retrun_frequecy_dict
```

```python
nanatmaj_hpid_freq_dict = calculte_hpid_freq(combined_nanatmaj_phenotypes_df, hpo_disease_dict)
nanatmaj_hpid_freq_df = pd.DataFrame.from_dict(nanatmaj_hpid_freq_dict, orient= 'index')
nanatmaj_hpid_freq_df.columns = ['Phenotype_ID', 'Frequency']
```
