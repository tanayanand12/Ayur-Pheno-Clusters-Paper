```python
def fetch_hpids(disease_id):
  temp_hpid_name_dict = defaultdict(list)
  hpid_list = disease_hpo_dict[disease_id]
  for i in hpid_list:
    # print(i, '\t', hpid_descriptions_dict[i], '\n')
    # print(hpid_descriptions_dict[i], '\n')
    temp_hpid_name_dict[i].append(hpid_descriptions_dict[i])
  # print(temp_hpid_name_dict)
  return temp_hpid_name_dict
```

```python
def fetch_disease_name(disease_id):
  disease_name = disease_name_dict[disease_id]
  return disease_name
```

```python
# test
# type(fetch_disease_name('OMIM:301040'))
```

```python
def fetch_hpid_descriptions(hpid):
  hpid_description = hpid_descriptions_dict[hpid]
  return hpid_description
```

```python
def fetch_hpid_vpk_label(hpid):
  hpid_vpk_label_list = hpid_vpk_label_dict[hpid]
  # print(hpid_vpk_label_list)
  hpid_vpk_label_list.insert(0, hpid)

  phenotype_name = fetch_hpid_descriptions(hpid)
  hpid_vpk_label_list.insert(1, phenotype_name[0])

  # print(hpid_vpk_label_list)

  hpid_vpk_label_df = pd.DataFrame(hpid_vpk_label_list).T
  # hpid_vpk_label_df.columns = ['Phenotype ID', 'Phenotype Name', 'V Label', 'P Label', 'K Label']
  hpid_vpk_label_list.clear()
  return hpid_vpk_label_df
```

```python
def fetch_corresp_diseases(hpid):
  disease_list = hpo_disease_dict[hpid]
  # print(disease_list)
  temp_disease_name_dict = defaultdict(list)
  for i in disease_list:
    temp_list = disease_name_dict[i]
    temp_disease_name_dict[i].append(temp_list[0])
  disease_name_df = pd.DataFrame.from_dict(temp_disease_name_dict).T
  # disease_name_df.index.name = None
  disease_name_df = disease_name_df.reset_index()
  return disease_name_df
```

```python
def fetch_disease_vpk_count(hpid):
  disease_vpk_count_list = disease_vpk_count_dict[hpid]
  disease_vpk_count_list.insert(0, hpid)

  disease_name = fetch_disease_name(hpid)
  disease_vpk_count_list.insert(1, disease_name[0])

  # print(disease_vpk_count_list)

  disease_vpk_count_df = pd.DataFrame(disease_vpk_count_list).T
  disease_vpk_count_list.clear()
  disease_vpk_count_df.columns = ['Disease ID', 'Disease Name', 'V Count', 'P Count', 'K Count']
  return disease_vpk_count_df
```

```python
def disease_details(disease_id):
  a = disease_id
  hpid_list = fetch_hpids(a) # dict id to name mapping
  disease_name = fetch_disease_name(a)
  disease_vpk_count_list = fetch_disease_vpk_count(a)
  return hpid_list, disease_name, disease_vpk_count_list
```

```python
# Code for Disease Information Retrival

def disease_info_retrive(disease_id_list):
  n = len(disease_id_list)

  main_hpid_list_dataframe = pd.DataFrame()
  disease_name_list  = []
  main_disease_vpk_count_dataframe = pd.DataFrame(columns=['Disease ID', 'Disease Name', 'V Count', 'P Count', 'K Count'])

  for disease_id in disease_id_list:
    hpids_list, disease_name, disease_vpk_count = disease_details(disease_id)

    hpid_list_df = pd.DataFrame.from_dict(hpids_list).T

    main_hpid_list_dataframe = pd.concat([main_hpid_list_dataframe, hpid_list_df], axis=0)

    disease_name_list.append(disease_name)

    main_disease_vpk_count_dataframe = pd.concat([main_disease_vpk_count_dataframe, disease_vpk_count], axis=0)

  # main_hpid_list_dataframe.columns= ['Phenotype_ID', 'Phenotype_Name']

  #drop duplicates
  main_disease_vpk_count_dataframe.drop_duplicates()
  main_hpid_list_dataframe.drop_duplicates()

  return main_hpid_list_dataframe, disease_name_list, main_disease_vpk_count_dataframe
```
