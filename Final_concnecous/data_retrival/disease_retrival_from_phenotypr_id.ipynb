```python
def pickle_load(file_path):
    with open(file_path, 'rb') as handle:
        b = pickle.load(handle)
    return b
```

```python
vata_phenotypes_df = pd.read_excel(r'Vata_phenotypes.xlsx')
vata_phenotypes_df.drop_duplicates()
```

```python
vata_phenotypes_df_list = vata_phenotypes_df.values.tolist()
vata_phenotypes_df_list = [i for row in vata_phenotypes_df_list for i in row]
```

```python
main_disease_list_vata = []
res_hpid_column_list = []
suspicious_pid_list = []

# count = 0
for p_id in vata_phenotypes_df['Phenotype_ID']:
    temp_disease_list  = hpo_disease_dict[p_id]
    if (temp_disease_list == []):
        suspicious_pid_list.append(p_id)
    # temp_hpid_list = np.full((len(temp_disease_list)), p_id)
    temp_hpid_list = []
    given_value= p_id
    temp_hpid_list.extend([given_value for i in range(len(temp_disease_list))])
    # print('len of disease list: ', len(temp_disease_list))
    # print("length of temp: ",len(temp_hpid_list))
    main_disease_list_vata.append(temp_disease_list)
    res_hpid_column_list.append(temp_hpid_list)
    # print('length of res: ',len(res_hpid_column_list), "\n\n\n")
    # print(res_hpid_column_list)
    # if (count == 10):
    #     break
    
    # count += 1

temp_disease_list.clear()
temp_hpid_list.clear()

main_disease_list_vata = [i for row in main_disease_list_vata for i in row]
res_hpid_column_list = [i for row in res_hpid_column_list for i in row]
# main_disease_list_vata = np.unique(main_disease_list_vata)
len(main_disease_list_vata)
```
